# Time column

```python
# Split to Hour and Minutes
df['time'] = df['time'].str.strip()
df[['hour', 'minutes']] = df['time'].str.split(' and ', expand=True)
df
```

### Hour column

```python
# Filter for rows that end with hrs
endswith_hrs = df['hour'].str.endswith('hrs')
df[endswith_hrs]
```

```python
# Filter for other pattern
df[~df['hour'].str.endswith('hrs')]
```

```python
# Filter for rows that end with hr
endswith_hr = df['hour'].str.endswith('hr')
df[endswith_hr]
```

```python
# Filter for other rows that not end with hrs or hr
df[~(df['hour'].str.endswith('hrs') | (df['hour'].str.endswith('hr')))]
```

```python
# Filter for rows that end with mins
endswith_mins = df['hour'].str.endswith('mins')
df[endswith_mins]
```

```python
# Filter for other rows that not end with hrs, hr, or mins
df[~(df['hour'].str.endswith('hrs') | df['hour'].str.endswith('hr') | df['hour'].str.endswith('mins'))]
```

```python
# Filter for rows that end with min
endswith_min = df['hour'].str.endswith('min')
df[endswith_min]
```

```python
# Filter for rows that equal to Less than 1 minute
lessthan_one_min = df['hour'] == 'Less than 1 minute'
df[lessthan_one_min]
```

```python
# To check that the number of rows of all pattern is equivalent to the number of rows of df
assert len(df[endswith_hrs] + df[endswith_hr] + df[endswith_mins] + df[endswith_min] + df[lessthan_one_min]) == len(df), \
    "The number of rows do not consistent"
```

```python
# Convert all value into minutes then cast to INTEGER type
df['hour'] = df['hour'].apply(lambda x: int(x.replace(' hrs', '').strip()) * 60 if x.endswith('hrs')
                                        else int(x.replace(' hr', '').strip()) * 60 if x.endswith('hr')
                                        else int(x.replace(' mins', '').strip()) if x.endswith('mins')
                                        else int(x.replace(' min', '').strip()) if x.endswith('min')   
                                        else 1 if x == 'Less than 1 minute'
                                        else x)

df
```

### Minutes column

```python
# Replace NULL with STRING 0
df['minutes'] = df['minutes'].fillna('0')
df
```

```python
# Filter for rows that end with mins
endswith_mins = df['minutes'].str.endswith('mins')
df[endswith_mins]
```

```python
# Filter for other pattern
df[~endswith_mins]
```

```python
# Filter for rows that end with min
endswith_min = df['minutes'].str.endswith('min')
df[endswith_min]
```

```python
# Filter for rows that not end with mins or min
df[~(endswith_mins | endswith_min)]
```

```python
# Filter for rows that is equal to 0
zero = df['minutes'] == '0'
df[zero]
```

```python
# To check that the number of rows of all pattern is equivalent to the number of rows of df
assert len(df[endswith_mins] + df[endswith_min] + df[zero]) == len(df), \
    "The number of rows do not consistent"
```

```python
# Cast to INTEGER type
df['minutes'] = df['minutes'].apply(lambda x: int(x.replace(' mins', '').strip()) if x.endswith('mins')
                                        else int(x.replace(' min', '').strip()) if x.endswith('min')   
                                        else 0 if x == '0'
                                        else x)

df
```

```python
# Create new column named 'duration_mins', hour + minutes
df['duration_mins'] = df['hour'] + df['minutes']
df
```
