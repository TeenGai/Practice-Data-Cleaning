# Import file

```python
import pandas as pd
df = pd.read_csv('raw_cricket.csv')
df
```

```python
df.info()
```

# Drop unnecessary columns and Rename

Columns to keep to solve the problem
- Player 
- Span 
- Inns 
- HS 
- SR 
- 100 
- 50 
- 0

```python
df = df.drop(['Mat', 'NO', 'Runs', 'Ave', 'BF', '4s', '6s'], axis=1)
df
```

```python
df = df.rename(columns={
    'HS':'Highest_Inns_Score', 
    'SR':'Batting_Strike_Rate', 
    '100':'Hundreds', 
    '50':'Fifties', 
    '0':'Ducks'
})
df
```

# Drop duplicates

```python
df.duplicated().sum()
```

```python
df[df.duplicated(keep=False)]
```

```python
drop_dup = df.drop_duplicates().reset_index(drop=True)

# to guarantee that the number of rows after dropping duplicates is correct
# Before drop - 66 rows, After drop must be 62 rows
assert len(drop_dup) == len(df) - df.duplicated().sum(), \
"The number of rows after dropping duplicates are not consistent"

df = drop_dup
df
```

# Check null

```python
df.isna().sum()
```

```python
df[df['Batting_Strike_Rate'].isna()]
```

```python
# Replace Null with Average
df['Batting_Strike_Rate'] = df['Batting_Strike_Rate'].fillna(df['Batting_Strike_Rate'].mean().round(2))

# Recheck the result
df.iloc[[8, 14, 54], :]
```

# Format Player

```python
country_pattern = r'.*\s\([A-Z/]+\)$'
assert len(df['Player'].str.match(country_pattern) == len(df)), \
    "There is some rows not match with the pattern provided"

df
```

```python
df['Country'] = df['Player'].str.split('(').str[1] \
                                .str.split(')').str[0] \
                                    .str.strip()

df['Player'] = df['Player'].str.split('(').str[0] \
                                .str.strip()

df
```

```python
df.dtypes
```

# Split Span into Start_Career and Retire_Career

```python
df['Span'] = df['Span'].str.strip()
```

```python
span_pattern = r'^\d{4}-\d{4}$'

assert len(df['Span'].str.match(span_pattern)) == len(df), \
    "There is some rows not match with the pattern provided"

df
```

```python
df[['Start_Career', 'Retire_Career']] = df['Span'].str.split('-', expand=True)
df
```

```python
df['Start_Career'] = df['Start_Career'].astype(int)
df['Retire_Career'] = df['Retire_Career'].astype(int)
df.dtypes
```

```python
df = df.drop('Span', axis=1)
df
```

# Format Highest_Inns_Score

```python
HIS_pattern = r"^(\d+)$"
df[df['Highest_Inns_Score'].str.match(HIS_pattern)]
```

```python
df[~df['Highest_Inns_Score'].str.match(HIS_pattern)]
```

```python
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].str.extract(r'(\d+)')
df
```

```python
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].astype(int)
df.dtypes
```

# Question 1 -> What is the Average Career Length

```python
df['Career_Length'] = (df['Retire_Career'] - df['Start_Career'])
df['Career_Length'].mean()
```

# Question 2 -> AVG Batting_Strike_Rate for cricketers who played over 10 years

```python
df.loc[df['Career_Length'] > 10, 'Batting_Strike_Rate'].mean()
```

# Question 3 -> Find number of cricketers who played before 1960

```python
df.loc[df['Start_Career'] < 1960, 'Player'].count()
```

# Question 4 -> Max Highest Inns Score by Country

```python
df.groupby('Country')['Highest_Inns_Score'].max().sort_values(ascending=False)
```

# Question 5 -> Hundreds, Fifties, Ducks (0) AVG by Country

```python
df.groupby('Country')[['Hundreds', 'Fifties', 'Ducks']].mean()
```
